# Lab04.4  การวาดแผนภาพตามแนวคิด Association abstraction

Association abstraction เป็นการแสดงความสัมพันธ์ระหว่าง class ที่มีความสัมพันธ์แบบเกี่ยวพันกัน ไม่สามารถอธิบายโดย Abstraction แบบอื่น ๆ ได้
- ไม่ใช่ "Is a" แบบ Classification
- ไม่ใช่ "Is part of" แบบ Aggregation
- ไม่ใช่ "Is kind of" แบบ Generalization
- แต่เป็น "Is related to" 

### รูปแบบ
``` plantuml
@startuml
"Class A" -- "Class B" :  relationship
"Class C" -- "Class D" : > relationship
"Class E" -- "Class F" : < relationship
@enduml
```

![](./Pictures/Pict-35.png)

### ตัวอย่าง

1. คนเป็นเจ้าของรถยนต์

``` plantuml
@startuml
class Car
class Person
Car -- Person : < owns
@enduml
```

## Cardinality ใน Association Abstraction

“Cardinality”  คือตัวเลขที่ใช้แสดงจำนวนของสมาชิกที่สามารถมีได้ใน Class หนึ่งๆ ที่มีส่วนร่วมใน Association

### ตัวอย่าง 
ผู้หญิง (แม่) สามารถมีลูกได้ตั้งแต่ 0 คน ถึงกี่คนก็ได้ ในทางกลับกัน ลูก 1 คน สามารถมีแม่ได้เพียงคนเดียว

![](./Pictures/Pict-37.png)




ขั้นตอนที่ 1 : เขียน class 2 class ที่มีความสัมพันธ์และลากเส้นตรงใส่ชื่อแสดงความสัมพันธ์

``` plantuml

{
    - Attribues
    + Method()
``` 

### หมายเหตุ

เราสามารถควบคุมทิศทางการวาดคลาสไดอะแกรมได้โดยการเปลี่ยนเครื่องหมาย `-` เป็น `--` ดังตัวอย่าง

โดยจำนวนที่เพิ่มขึ้นของ `-` จะมีส่วนเพิ่มระยะห่างระหว่างคลาสด้วย 


``` plantuml
@startuml
Mom "1..1" - "0..n" Son : > has
Person "1..1" -- "0..n" Car : > has
Forest "1..1" --- "0..n" tree : > has
@enduml
```

จะได้ผลลัพธ์ดังรูป

![](./Pictures/Pict-43.png)






##  แบบฝึกหัด

1. จงเขียนความสัมพันธ์ในโลกของความเป็นจริง ให้เป็น association diagram โดยเพิ่ม cardinality และ กำหนดทิศทางความสัมพันธ์ให้ถูกต้อง

   1.1 คนเป็นเจ้าของรถยนต์
   
   1.2 แม่มีลูก
   
   1.3 สามีรักภรรยา
   
   1.4 ดินสออยู่ในกระเป๋า
   
   1.5 นักการเมืองออกกฎหมาย
   
   1.6 ทหารใส่เครื่องแบบ
   
   1.7 ประธานบริษัทบริหารกิจการ
   
   1.8 กระดานดำอยู่ในห้องเรียน

1.1 คนเป็นเจ้าของรถยนต์
Entity: คน (Person) → รถยนต์ (Car)

Relationship: เป็นเจ้าของ (Owns)

Cardinality: 1 คน → หลายรถยนต์ (1:M) | รถยนต์ 1 คัน → 1 คน (M:1)

ทิศทาง: "คนเป็นเจ้าของรถยนต์" → ลูกศรชี้จาก คน ไปที่ รถยนต์

(Person) ── Owns ──> (Car)

1.2 แม่มีลูก
Entity: แม่ (Mother) → ลูก (Child)

Relationship: มี (Has)

Cardinality: 1 แม่ → หลายลูก (1:M) | 1 ลูก → 1 แม่ (M:1)

ทิศทาง: "แม่มีลูก" → ลูกศรชี้จาก แม่ ไปที่ ลูก

(Mother) ── Has ──> (Child)

1.3 สามีรักภรรยา
Entity: สามี (Husband) ↔ ภรรยา (Wife)

Relationship: รัก (Loves)

Cardinality: โดยทั่วไปเป็น 1:1

ทิศทาง: ความสัมพันธ์เป็น สองทาง (Bidirectional)

(Husband) ←── Loves ──> (Wife)

1.4 ดินสออยู่ในกระเป๋า
Entity: ดินสอ (Pencil) → กระเป๋า (Bag)

Relationship: อยู่ใน (Is in)

Cardinality: 1 กระเป๋า → หลายดินสอ (1:M) | 1 ดินสอ → 1 กระเป๋า (M:1)

ทิศทาง: "ดินสออยู่ในกระเป๋า" → ลูกศรชี้จาก ดินสอ ไปที่ กระเป๋า

(Pencil) ── Is in ──> (Bag)

1.5 นักการเมืองออกกฎหมาย
Entity: นักการเมือง (Politician) → กฎหมาย (Law)

Relationship: ออก (Creates)

Cardinality: 1 นักการเมือง → หลายกฎหมาย (1:M) | 1 กฎหมาย → หลายนักการเมือง (M:M)

ทิศทาง: "นักการเมืองออกกฎหมาย" → ลูกศรชี้จาก นักการเมือง ไปที่ กฎหมาย

(Politician) ── Creates ──> (Law)

1.6 ทหารใส่เครื่องแบบ
Entity: ทหาร (Soldier) → เครื่องแบบ (Uniform)

Relationship: ใส่ (Wears)

Cardinality: 1 ทหาร → หลายเครื่องแบบ (1:M) | 1 เครื่องแบบ → หลายทหาร (M:M)

ทิศทาง: "ทหารใส่เครื่องแบบ" → ลูกศรชี้จาก ทหาร ไปที่ เครื่องแบบ

(Soldier) ── Wears ──> (Uniform)

1.7 ประธานบริษัทบริหารกิจการ
Entity: ประธานบริษัท (CEO) → กิจการ (Business)

Relationship: บริหาร (Manages)

Cardinality: 1 ประธานบริษัท → หลายกิจการ (1:M) | 1 กิจการ → 1 ประธานบริษัท (M:1)

ทิศทาง: "ประธานบริษัทบริหารกิจการ" → ลูกศรชี้จาก ประธานบริษัท ไปที่ กิจการ

(CEO) ── Manages ──> (Business)

1.8 กระดานดำอยู่ในห้องเรียน
Entity: กระดานดำ (Blackboard) → ห้องเรียน (Classroom)

Relationship: อยู่ใน (Is in)

Cardinality: 1 ห้องเรียน → หลายกระดานดำ (1:M) | 1 กระดานดำ → 1 ห้องเรียน (M:1)

ทิศทาง: "กระดานดำอยู่ในห้องเรียน" → ลูกศรชี้จาก กระดานดำ ไปที่ ห้องเรียน

(Blackboard) ── Is in ──> (Classroom)

![สกรีนช็อต 2025-03-23 155047.png](<attachment:สกรีนช็อต 2025-03-23 155047.png>)
